### Import Packages

```python
from selenium import webdriver
import pandas as pd 
```

### Open Link

```python
driver = webdriver.Chrome('C:\WebDriver\chromedriver.exe')

driver.get('https://pokemondb.net/pokedex/all')
driver.maximize_window()
```

```python
# Policy Policy
button = driver.find_element_by_class_name('gdpr-accept')

try:
    button.click()
except:
    pass
```
